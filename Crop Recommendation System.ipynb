```python
import tkinter as tk
from tkinter import ttk
#from PIL import Image, ImageTk
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Load dataset
data = pd.read_csv("Crop_recommendation.csv")

# Create the model
features = data[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']]
target = data['label']
RF = RandomForestClassifier(n_estimators=20, random_state=0)
RF.fit(features, target)

# Create the GUI
root = tk.Tk()
root.title("Crop Recommendation System")
root.geometry("500x500")
root.configure(bg="white")

#Load and resize the background image
bg_image = tk.PhotoImage(file="C:/Users/dell/OneDrive/Desktop/DAA/b.png")
#bg_image = bg_image.resize((200, 200), Image.ANTIALIAS)
my_label=tk.Label(root, image=bg_image)
my_label.place(x=0,y=0, relwidth=1, relheight=1)





# Create the labels and entries for user inputs
n_label = tk.Label(root, text="Nitrogen (N)", font=("Helvetica", 14), bg="white")
n_label.place(relx=0.2, rely=0.1, anchor="w")
n_entry = ttk.Entry(root, font=("Helvetica", 14), width=10)
n_entry.place(relx=0.6, rely=0.1, anchor="w")

p_label = tk.Label(root, text="Phosphorus (P)", font=("Helvetica", 14), bg="white")
p_label.place(relx=0.2, rely=0.2, anchor="w")
p_entry = ttk.Entry(root, font=("Helvetica", 14), width=10)
p_entry.place(relx=0.6, rely=0.2, anchor="w")

k_label = tk.Label(root, text="Potassium (K)", font=("Helvetica", 14), bg="white")
k_label.place(relx=0.2, rely=0.3, anchor="w")
k_entry = ttk.Entry(root, font=("Helvetica", 14), width=10)
k_entry.place(relx=0.6, rely=0.3, anchor="w")

temp_label = tk.Label(root, text="Temperature (Â°C)", font=("Helvetica", 14), bg="white")
temp_label.place(relx=0.2, rely=0.4, anchor="w")
temp_entry = ttk.Entry(root, font=("Helvetica", 14), width=10)
temp_entry.place(relx=0.6, rely=0.4, anchor="w")

hum_label = tk.Label(root, text="Humidity (%)", font=("Helvetica", 14), bg="white")
hum_label.place(relx=0.2, rely=0.5, anchor="w")
hum_entry = ttk.Entry(root, font=("Helvetica", 14), width=10)
hum_entry.place(relx=0.6, rely=0.5, anchor="w")

ph_label = tk.Label(root, text="pH", font=("Helvetica", 14), bg="white")
ph_label.place(relx=0.2, rely=0.6, anchor="w")
ph_entry = ttk.Entry(root, font=("Helvetica", 14), width=10)
ph_entry.place(relx=0.6, rely=0.6, anchor="w")

rain_label = tk.Label(root, text="Rainfall (mm)", font=("Helvetica", 14), bg="white")
rain_label.place(relx=0.2, rely=0.7, anchor="w")
rain_entry = ttk.Entry(root, font=("Helvetica", 14), width=10)
rain_entry.place(relx=0.6, rely=0.7, anchor="w")

# Define the function to Create the labels for displaying the predicted crop

result_label = tk.Label(root, text="", font=("Helvetica", 20), bg="white")
result_label.place(relx=0.5, rely=0.8, anchor="center")

# Define the function to get the predicted crop
def get_prediction():
    # Get the user inputs
    N = float(n_entry.get())
    P = float(p_entry.get())
    K = float(k_entry.get())
    temp = float(temp_entry.get())
    humidity = float(hum_entry.get())
    ph = float(ph_entry.get())
    rainfall = float(rain_entry.get())
    
    # Make the prediction using the model
    prediction = RF.predict([[N, P, K, temp, humidity, ph, rainfall]])
    crop = prediction[0]
    
    # Display the predicted crop
    result_label.config(text=f"Predicted Crop: {crop}")

# Create the button to make the prediction
predict_button = tk.Button(root, text="Predict", font=("Helvetica", 14), bg="lightgreen", command=get_prediction)
predict_button.place(relx=0.5, rely=0.9, anchor="center")

root.mainloop()
```
